# Expérience 2 : Géophysique

Protocole de laboratoire [disponible ici](https://sitescours.monportail.ulaval.ca/contenu/sitescours/036/03613/202009/site121785/modules809962/module1031407/page2793321/bloccontenu2672853/GPH-3000-A2019-g%C3%A9ophysique.pdf?identifiant=eaedfc8ad9ff64adbd26db02f6da12c05f4fcac0).

## Préparation

**Date de préparation :** 23 septembre 2020

**Date de l'expérimentation :** 24 septembre 2020

**Coéquipier :** Pierre-Olivier Janvier

**But :** Étudier les phénomènes de la propagation des ondes dans le sol afin de pouvoir se renseigner sur son contenu, et ce, à l'aide d'une expérience de «sismique-réfraction» utilisant des ondes P (ondes longitudinales).

**Objectifs spécifiques :**
- Vérifier le bon fonctionnement d'un géophone.
- Déterminer la fréquence de résonance du géophone.
- Étalonner le géophone (amplitude des vibrations en fonction de la tension de sortie).
- Concevoir un système d'acquisition afin d'enregistrer les signaux d'un ou de plusieurs géophones.
- Mesurer l'épaisseur du mort-terrain sous l'université Laval.

**Matériel nécessaire :**
- Géophone
- Oscilloscope
- Table vibrante
- Générateur de fonction
- Capteur de position calibré [LD-300-15](https://br.omega.com/omegaFiles/pressure/pdf/LD300.pdf)
- Carte d'aquisition [USB-6008](https://www.ni.com/pdf/manuals/375295c.pdf)
- système d'acquisition [Labjack U3-LV](https://files.labjack.com/datasheets/LabJack-U3-Datasheet.pdf)
- masse manuelle

**Conception :**
- Pour ce qui est de la détermination de la fréquence de résonance du géophone, les deux méthodes utilisées seront en trouvant la fréquence de résonance avec le générateur de fonction et la table vibrante comme illustré à la figure 1 et en observant la réponse à une impulsion de celui-ci. Dans la première méthode, la fréquence de résonance se trouvera à la fréquence du générateur lorsque les oscillations de sorties seront les plus grandes possibles et pour la deuxième méthode, la fréquence sera celle des oscillations présentes lors du retour au point d'équilibre de l'oscillateur après avoir reçu une impulsion.

- Pour ce qui est de la prise de mesure sur le terrain, la prise de donnée se fera avec la carte d'acquisition de NI USB-6008 et le VI (geophys.vi) déjà donné avec le laboratoire par l'intermédiaire du site de cours. Le guide d'utilisation du VI est accessible [ici](https://sitescours.monportail.ulaval.ca/contenu/sitescours/036/03613/202009/site121785/modules809962/module1031407/page2793321/bloccontenu2672854/userguide-geophys_VI.pdf?identifiant=907d26de0eb3bbcc1de07222d88f7ec6595ab1e4).
